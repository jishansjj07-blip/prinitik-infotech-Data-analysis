```python
#Content Type vs Engagement
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Post_Type", y="Engagement")
plt.title("Content Type vs Engagement")
plt.xlabel("Content Type")
plt.ylabel("Average Engagement")
plt.show()
```

```python
#Platform vs Engagement
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Platform", y="Engagement")
plt.title("Platform vs Engagement")
plt.xlabel("Platform")
plt.ylabel("Average Engagement")
plt.show()
```

```python
#Posting Frequency
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Platform")
plt.title("Posting Frequency by Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Posts")
plt.show()
```

```python
#Engagement Trend
df["Date"] = pd.to_datetime(df["Date"])

daily_engagement = df.groupby("Date")["Engagement"].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=daily_engagement, x="Date", y="Engagement", marker="o")
plt.title("Engagement Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Average Engagement")
plt.xticks(rotation=45)
plt.show()
```

## Insights

2. Platform X recorded the highest average engagement (approximately 3127.69), indicating stronger audience interaction.

3. LinkedIn generated the highest total engagement because it had the highest number of published posts.

4. Reel posts produced the highest total engagement, showing consistent performance.

5. The top-performing posts achieved engagement values above 5,500, indicating that some posts attracted significantly higher audience interaction.

6. Likes have the strongest positive correlation with Engagement, meaning they contribute the most to the overall engagement score.

7. Comments and Shares have a weaker correlation with Engagement compared to Likes.

8. Engagement varied across different posting dates, highlighting the importance of posting time.
